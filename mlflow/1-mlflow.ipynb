```python
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import mlflow
from mlflow.models import infer_signature
```

```python
### set tracking uri
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")
```

```python
X,y=datasets.load_iris(return_X_y=True)
## split the data in traininng and tst
X_train,X_text,y_train,y_test=train_test_split(X,y,test_size=0.2)

### Define the model hyperparameter
params={"penalty":"l2","solver":"lbfgs","max_iter":1000,"multi_class":"auto","random_state":8888}

lr=LogisticRegression(**params)
lr.fit(X_train,y_train)
```

```python
### Prediction on test data
y_pred=lr.predict(X_text)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
### MLFlow tracking
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")

### Create a new MLFlow experiment
mlflow.set_experiment("MLFLOW Quickstart")

## start an mlflow run
with mlflow.start_run():
    ### log the hyperparameter
    mlflow.log_params(params)

    ### Log the accuraccy matrix
    mlflow.log_metric("accuracy",accuracy)
    ## set tags that we can use to remind ourselfs what this run was for
    mlflow.set_tag("Training info","Basic LR model for IRIS data")

    ### infer the model signature
    ### signature represents model input and output as data frames with optionally named columns and data type specified as one of types in defined in mlflow.types.Datatype
    ### It also includes parameters scheama for inference. This mehtod will raise and exception if the data contains incompatible types 
    signature=infer_signature(X_train,lr.predict(X_train))

    ## Log the model
    model_info=mlflow.sklearn.log_model(sk_model=lr,artifact_path="iris_model",
                                        signature=signature,input_example=X_train,
                                        registered_model_name="tracking-quickstart")
```

```python
X,y=datasets.load_iris(return_X_y=True)
## split the data in traininng and tst
X_train,X_text,y_train,y_test=train_test_split(X,y,test_size=0.2)

### Define the model hyperparameter
params={"penalty":"l2","solver":"newton-cg","max_iter":1000,"multi_class":"auto","random_state":8888}

lr=LogisticRegression(**params)
lr.fit(X_train,y_train)
```

```python
### MLFlow tracking
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")

### Create a new MLFlow experiment
mlflow.set_experiment("MLFLOW Quickstart")

## start an mlflow run
with mlflow.start_run():
    ### log the hyperparameter
    mlflow.log_params(params)

    ### Log the accuraccy matrix
    mlflow.log_metric("accuracy",accuracy)
    ## set tags that we can use to remind ourselfs what this run was for
    mlflow.set_tag("Training info","Basic LR model for IRIS data")

    ### infer the model signature
    ### signature represents model input and output as data frames with optionally named columns and data type specified as one of types in defined in mlflow.types.Datatype
    ### It also includes parameters scheama for inference. This mehtod will raise and exception if the data contains incompatible types 
    signature=infer_signature(X_train,lr.predict(X_train))

    ## Log the model
    model_info=mlflow.sklearn.log_model(sk_model=lr,artifact_path="iris_model",
                                        signature=signature,input_example=X_train,
                                        registered_model_name="tracking-quickstart")
```

```python
model_info.model_uri
```

#### Load the entire model in generic python function model

```python
loaded_model=mlflow.pyfunc.load_model(model_info.model_uri)
```

```python
predictions=loaded_model.predict(X_text)
iris_features_name=datasets.load_iris().feature_names
result=pd.DataFrame(X_text,columns=iris_features_name)
```
